```python
# Importing the CSV file to have a look at the first few rows
import pandas as pd

# Loading the data
file_path = '/content/positionnement_politique_v3.csv'
data = pd.read_csv(file_path)

# Displaying the first 5 rows of the dataframe
data.head()
```

```python
# Importing the required libraries for visualization
import matplotlib.pyplot as plt

# Getting the counts of candidates for each political positioning
positioning_counts = data['positionnement'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
positioning_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Taux de positionnement des candidats')
plt.ylabel('') # Removing the y-axis label for a cleaner look
plt.show()

# Returning the percentages for further reference
positioning_counts_percentage = positioning_counts / positioning_counts.sum() * 100
positioning_counts_percentage
```

```python
# Importing the CSV file to have a look at the first few rows
import pandas as pd

# Loading the data
file_path = '/content/resultats-par-niveau-cirlg-t1-france-entiere.csv'
data2 = pd.read_csv(file_path)

# Displaying the first 5 rows of the dataframe
data2.head()
```

```python
# Grouping the data by department and summing the relevant columns
department_data2 = data2.groupby('Libellé du département')[['Inscrits', 'Votants', 'Abstentions']].sum()

# Plotting the bar graph for the first 10 departments (for visibility)
department_data2.head(10).plot(kind='bar', figsize=(12, 6))
plt.title('Inscrits, Votants et Abstentions par Département (10 premiers départements)')
plt.xlabel('Département')
plt.ylabel('Nombre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Selecting the first constituency of the department "Ain"
first_constituency_ain = data2[data2['Libellé de la circonscription'] == '1ère circonscription']

# Extracting the votes for each candidate (we will take the first 5 candidates for simplicity)
votes_candidates = first_constituency_ain[['Nom_1', 'Voix_1', 'Nom_2', 'Voix_2', 'Nom_3', 'Voix_3', 'Nom_4', 'Voix_4', 'Nom_5', 'Voix_5']]
votes_candidates = votes_candidates.melt(id_vars=['Nom_1', 'Nom_2', 'Nom_3', 'Nom_4', 'Nom_5'], value_vars=['Voix_1', 'Voix_2', 'Voix_3', 'Voix_4', 'Voix_5'])
votes_candidates['Candidate'] = votes_candidates.apply(lambda row: row['Nom_' + row['variable'].split('_')[-1]], axis=1)
votes_candidates = votes_candidates.groupby('Candidate')['value'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
votes_candidates.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Répartition des voix pour les principaux candidats\n1ère circonscription, Ain')
plt.ylabel('') # Removing the y-axis label for a cleaner look
plt.show()
```

```python
# Calculating the abstention percentage for each department
department_data2['% Abstention'] = department_data2['Abstentions'] / department_data2['Inscrits'] * 100

# Plotting the line graph for the abstention percentage
department_data2['% Abstention'].plot(kind='line', figsize=(12, 6))
plt.title('Pourcentage d\'Abstention par Département')
plt.xlabel('Département')
plt.ylabel('Pourcentage d\'Abstention')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

TOP 10 ABSTENTION

```python
# Sorting the department data by the abstention percentage in descending order
top_10_abstention = department_data2['% Abstention'].sort_values(ascending=False).head(10)

# Displaying the top 10 departments with the highest abstention rate
top_10_abstention
```

```python
# Extracting the last names from the candidates_positioning dataframe and converting to uppercase
data['nom_candidat'] = data['nom_candidat'].apply(lambda name: name.split()[-1].upper())

# Merging the votes data with the candidates' positioning again
merged_data = pd.merge(data2, data, left_on='nom_candidat', right_on='nom_candidat', how='left')

# Grouping the data by department and positioning details to get the vote percentages
vote_percentages_by_positioning = merged_data.groupby(['Département', 'Détails Positionnement'])['Pourcentage de Voix'].sum().unstack().reset_index()
vote_percentages_by_positioning = vote_percentages_by_positioning.fillna(0)

# Keeping only the required columns
vote_percentages_by_positioning = vote_percentages_by_positioning[['Département', 'extrême droite', 'droite', 'gauche', 'extrême gauche']]

# Merging with the top 10 abstention departments
final_result = pd.merge(top_10_abstention.reset_index(), vote_percentages_by_positioning, left_on='Libellé du département', right_on='Département', how='left')

# Displaying the result
final_result.drop(columns=['Département'], inplace=True)
final_result.columns = ['Département', 'Taux d\'Abstention (%)', 'Extrême Droite (%)', 'Droite (%)', 'Gauche (%)', 'Extrême Gauche (%)']
final_result
```
